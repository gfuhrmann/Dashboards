```python
# Agrupando pela coluna "population" a  média de "total_rooms" 
# Qual a média de total rooms que apresentam população igual a X?

db_1.groupby("population")["total_rooms"].mean()
```

```python
# Agrupando por várias variáveis estatísticas de uma só vez

db_1.groupby("population")["total_rooms"].agg([min, max, sum])
```

```python
# Pivot Tables (Tabelas Dinâmicas)
# Cria uma tabela Dinâmica com os valores(variável calculada: mediana) "population" e os índices de linha "households"

db_1.pivot_table(values="population", index = "households", aggfunc=np.median)
```

```python
# Duas variáveis estatísticas calculadas: média e mediana

db_1.pivot_table(values="population", index = "households", aggfunc=[np.median, np.mean])
```

```python
# Tabelas Dinâmicas de Duas Variáveis

db_1.pivot_table(values="population", index = "households", columns="housing_median_age")

# Determino: Valores da tabela (values), índice (index) e títulos das colunas (columns)
```

```python
#Substitui NaN por 0 e adidiona uma coluna final com o somatório

db_1.pivot_table(values="population", index = "households", columns="housing_median_age", fill_value=0, margins=True)
```

```python
# Definindo uma coluna como índice (de linha)

db_1.set_index("housing_median_age")
```
